```python
#task-1
class RealNumber:
  def __init__(self, r=0):
    self.__realValue = r
  def getRealValue(self):
    return self.__realValue
  def setRealValue(self, r):
    self.__realValue = r
  def __str__(self):
    return 'RealPart: '+str(self.getRealValue())
    
class ComplexNumber(RealNumber):
  def __init__(self,r=1,comp=1):
    super().__init__(float(r))
    self.comp=float(comp)
  def __str__(self):
    return f'{super().__str__()}\nImaginary part: {self.comp}'

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
#task-2

class RealNumber:
  def __init__(self, number=0):
    self.number = number
  def __add__(self, anotherRealNumber):
    return self.number + anotherRealNumber.number
  def __sub__(self, anotherRealNumber):
    return self.number - anotherRealNumber.number
  def __str__(self):
    return str(self.number)

class ComplexNumber(RealNumber):
  def __init__(self,number,nmbr=None):
    super().__init__(number)
    self.nmbr=nmbr
  def __add__(self,other):
    a=super().__add__(other.number)
    b=self.nmbr + other.nmbr
    return ComplexNumber(a,b)
  def __sub__(self,other):
    a=super().__sub__(other.number)
    b=self.nmbr - other.nmbr 
    return ComplexNumber(a,b)
  def __str__(self):
    if self.nmbr>0:
      return str(self.number)+" + "+str(self.nmbr)+"i"
    if self.nmbr<0:
      return str(self.number)+" - "+str(self.nmbr*(-1))+"i"
    

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#task-4
class Fruit:
  def __init__(self, formalin=False, name=''):
    self.__formalin = formalin
    self.name = name
  def getName(self):
    return self.name
  def hasFormalin(self):
    return self.__formalin
class testFruit:
  def test(self, f):
    print('----Printing Detail----')
    if f.hasFormalin():
      print('Do not eat the',f.getName(),'.')
      print(f)
    else:
      print('Eat the',f.getName(),'.')
      print(f)

class Mango(Fruit):
  def __init__(self,n=True,name='Mango'):
    super().__init__(n,name)
  def __str__(self):
    if self.hasFormalin():
      return f"Mangos are bad for you"
    else:
      return f"Mangos are good for you"

class Jackfruit(Fruit):
  def __init__(self,n=False,name='Jackfruit'):
    super().__init__(n,name)
  def __str__(self):
    if self.hasFormalin():
      return f"Jackfruits are bad for you"
    else:
      return f"Jackfruits are good for you"
    

m = Mango()
j = Jackfruit()
t1 = testFruit()
t1.test(m)
t1.test(j)
```

```python
#task-5

class Exam:
  def __init__(self,marks):
    self.marks = marks
    self.time = 60
  def examSyllabus(self):
    return "Maths , English"
  def examParts(self):
    return "Part 1 - Maths\nPart 2 - English\n"


class ScienceExam(Exam):
  def __init__(self,marks,time,*crs):
    super().__init__(marks)
    self.time=time
    self.crs=list(crs)
    self.len=2+len(self.crs)
  def __str__(self):
    return f"Marks:{self.marks} Time:{self.time} minutes Number of Parts: {self.len}"

  def examSyllabus(self):
    m=super().examSyllabus()
    x=str(m)
    for i in range (len(self.crs)):
      x+=" , "+str(self.crs[i])
    return x

  def examParts(self):
    n=super().examParts()
    x=str(n)
    for i in range (len(self.crs)):
      a=3+i
      if i==(len(self.crs)-1):
        x+="Part "+str(a)+" - "+str(self.crs[i])
      else:
        x+="Part "+str(a)+" - "+str(self.crs[i])+"\n"
    return x

      


engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture =ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

```

```python
#task-6
class Shape3D:
  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius
  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius
  def __str__(self):
    return "Radius: "+str(self._radius)
class Sphere(Shape3D):
  def __init__(self,name = 'Default',radius = 0,height='No need'):
    super().__init__(name,radius)
    self._height = height
    print (f"Shape name: {self._name} , Area Formula: 4 * pi * r * r")
  def calc_surface_area(self):
    self._area=super().calc_surface_area()*2*self._radius
    return f", Height: {self._height} \nArea: {self._area}"
  def __str__(self):
    a=self.calc_surface_area()
    b=super().__str__()
    return b+a

class Cylinder(Shape3D):
  def __init__(self,name = 'Default',radius = 0,height='No need'):
    super().__init__(name,radius)
    self._height = height
    print (f"Shape name: {self._name} , Area Formula: 2 * pi * r * (r + h)")
  def calc_surface_area(self):
    self._area=super().calc_surface_area()*(self._radius+self._height)
    return f", Height: {self._height} \nArea: {self._area}"
  def __str__(self):
    a=self.calc_surface_area()
    b=super().__str__()
    return b+a



sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#task-7
class PokemonBasic:
  def __init__(self, name = 'Default', hp = 0,weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type
  def get_type(self):
    return 'Main type: ' + self.type
  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'
  def __str__(self):
    return "Name: " + self.name + ", HP: " +str(self.hit_point) + ", Weakness: " + self.weakness

class PokemonExtra(PokemonBasic):
  def __init__(self,name = 'Default', hp = 0,weakness = 'None',*othr):
    super().__init__(name, hp,weakness)
    self.othr=list(othr)
    self.type=self.othr[0]
  def get_type(self):
    a=super().get_type()
    if len(self.othr)>=2:
      b=", Secondary type:"+str(self.othr[1])
      return a+b
    else:
      return a
  def get_move(self):
    c=super().get_move()
    var1="Other move: "
    if len(self.othr)>=3:
      if type(self.othr[2])==tuple or type(self.othr[2])==list :
        for i in range (len(self.othr[2])):
          if i==(len (self.othr[2])-1):
            var1+=self.othr[2][i]
          else:
            var1+=self.othr[2][i]+", "
      else:
        var1+=self.othr[2]
      return c+"\n"+var1
    else:
      return c

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water','Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water','Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#task-10

class Department:
  def __init__(self, s):
    self.semester = s
    self.name = "Default"
    self.id = -1
  def student_info(self):
    print("Name:", self.name)
    print("ID:", self.id)
  def courses(self, c1, c2, c3):
    print("No courses Approved yet!")
class CSE(Department):
  def __init__(self,name,id,sesn):
    self.semester = sesn
    self.name = name
    self.id = id
  def student_info(self):
    a=super().student_info()
    return a
  def courses(self, c1=None, c2=None, c3=None):
    self.c1=c1
    self.c2=c2
    self.c3=c3
    if self.c1==None and self.c2==None and self.c3==None:
      super().courses()
    else:
      print(f"Courses Approved to this CSE student in {self.semester} semester :")
      l1=[self.c1,self.c2,self.c3]
      for i in l1:
        if i!=None:
          print(i)
class EEE(Department):
  def __init__(self,name,id,sesn):
    self.semester = sesn
    self.name = name
    self.id = id
  def student_info(self):
    a=super().student_info()
    return a
  def courses(self, c1=None, c2=None, c3=None):
    self.c1=c1
    self.c2=c2
    self.c3=c3
    if self.c1==None and self.c2==None and self.c3==None:
      super().courses()
    else:
      print(f"Courses Approved to this EEE student in {self.semester} semester :")
      l1=[self.c1,self.c2,self.c3]
      for i in l1:
        if i!=None:
          print(i)



s1 = CSE("Rahim", 16101328,"Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
