```python
#task-1
class Student:
  count=0
  def __init__(self,name,course,age,cg):
    self.name=name
    self.crs=course
    self.age=age
    self.cg=cg
    Student.count+=1
    self.id=Student.count
  
  def get_details(self):
    print("ID:",self.id,"\nName:",self.name,"\nDepartment:",self.crs,"\nAge:",self.age,"\nCgpa:",self.cg)
  @classmethod
  def from_String(cls,info):
    name,course,age,cg=info.split("-")
    return cls(name,course,age,cg)

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print('--------------------------------------')
print('5) Class variable is defined inside a class and outside of the method. It is independent to the objects\n and belongs to class. It gives same values for all object. On the other hand, instance variable is object dependent,\n it is called inside a method and it gives different values for each object.')
print("\n6)Instance variable are the common variables that can access their instanes. They do not need \ndecorator. Again, they need to have class instances and need self parameter. On the contrary,class method is\nbounded by class. They cannot access specific instance data. They need decorator and they can call other static method.\n")
```

```python
#task-2
class Assassin:
  count=0
  def __init__(self,name,nm):
    Assassin.count+=1
    self.name=name
    self.nmbr=nm
  def printDetails(self):
    print("Name:",self.name,'\nSuccess rate:',self.nmbr,"%",'\nTotal number of Assassin:',Assassin.count)
  @classmethod
  def failureRate(cls,nam,nb):
    rate=100-nb 
    return cls(nam,rate)
  @classmethod
  def failurePercentage(cls,hola,dont):
    rate=100-dont 
    return cls(hola,rate)
    
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#task-3
class Passenger:
  count=0
  def __init__(self,name,money=450):
    Passenger.count+=1
    self.name=name
    self.mny=money
  def set_bag_weight(self,dot):
    if dot<=20:
      self.mny+=0
    elif dot>=21 and dot<=50:
      self.mny+=50
    elif dot>50:
      self.mny+=100
  def printDetail(self):
    print("Name:",self.name)
    print("Bus fare:",self.mny,"Taka")

print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
#task-4
class Travel:
  count=0
  def __init__(self,name=None,country=None,time=1): #name=source, country=destination
    Travel.count+=1
    self.name=name
    self.plc=country
    self.time=time

  def set_time(self,tm):
    self.time=tm

  def set_destination(self,x):
    self.plc=x

  def  set_source(self,y):
    self.name=y

  def display_travel_info(self):
    new="{:.2f}".format(self.time*1.00)
    return f"Source: {self.name} \nDestination: {self.plc} \nFlight Time: {new}"

print('No. of Traveller =', Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print('No. of Traveller =', Travel.count)
```

```python
#task-5
from datetime import date
class Employee:
  def __init__(self,name,workingPeriod):
    self.name=name
    self.workingPeriod=workingPeriod
    
  @classmethod
  def employeeByJoiningYear(cls,name,year1):
    idk=date.today().year-year1
    return cls(name,idk)
    
  @staticmethod
  def experienceCheck(workingPeriod,gender):
    if gender.lower()=='male':
      if workingPeriod<3:
        return f'He is not experienced'
      else:
        return f'He is experienced'
    if gender.lower()=='female':
      if workingPeriod<3:
        return f'She is not experienced'
      else:
        return f'She is experienced'


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#task-6
class Laptop:
  laptopCount=0
  def __init__(self,name,count):
    self.name=str(name)
    self.count=count
    Laptop.laptopCount+=count

  @staticmethod
  def advantage():
    print("Laptops are portable")

  @staticmethod
  def resetCount():
    Laptop.laptopCount=0
  
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#task-7
class Cat:
    Number_of_cats=0

    def __init__(self,col,act):
        self.col=col
        self.act=act
        Cat.Number_of_cats+=1

    def changeColor(self,col1):
        self.col=col1
        
    def printCat(self):
        print(self.col,'cat is',self.act)    
    @classmethod
    def no_parameter(cls,col='White',act='sitting'):
        obj1=cls(col,act)
        return obj1
    @classmethod
    def first_parameter(cls,col1):
        obj2=cls(col1,'sitting')
        return obj2
    @classmethod
    def second_parameter(cls,act1):
        obj3=cls('Grey',act1)
        return obj3
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task-8

import math
class Cylinder:
  radius=5
  height=18
  def __init__(self,radi,heigh):
    print('Default radius=',Cylinder.radius,'and height=',Cylinder.height)
    self.radius=radi
    self.height=heigh
    Cylinder.radius=self.radius
    Cylinder.height=self.height
    print('Updated: radius=',Cylinder.radius,'and height=',Cylinder.height)
  
  @staticmethod
  def area(radi,heigh):
    ar=(2*math.pi*radi*radi)+(2*math.pi*radi*heigh)
    print("Area:",ar)

  @staticmethod
  def volume(radi,heigh):
    vlm=math.pi*radi*radi*heigh
    print("Volume:",vlm)
  
  @classmethod
  def swap(cls,hei,rad):
    obj=cls(rad,hei)
    return obj 
    
  @classmethod
  def changeFormat(cls,string):
    rad,hei=string.split("-")
    obj=Cylinder(float(rad),float(hei))
    return obj
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#task-10
class SultansDine:
  totalbranch=0
  totalsell=0
  li1=[]
  def __init__(self,plc):
    self.plc=plc
    self.Branch_sell=0
    SultansDine.totalbranch+=1

  def sellQuantity(self,quantity):
    if quantity < 10:
      self.Branch_sell = quantity * 300
    elif quantity < 20:
      self.Branch_sell = quantity * 350
    else:
      self.Branch_sell = quantity * 400
    SultansDine.totalsell+=self.Branch_sell
    
  def branchInformation(self):
    print('Branch Name:',self.plc)
    print('Branch Sell:',self.Branch_sell,'Taka')
    SultansDine.li1.append([self.plc,self.Branch_sell])

  @classmethod
  def details(cls):
    print('Total Number of branch(s):',SultansDine.totalbranch)
    print("Total sell:",SultansDine.totalsell,"taka")
  
    for i in SultansDine.li1:
      print(f'Branch Name: {i[0]}, Branch Sell: {i[1]} Taka')
      x=(i[1] /SultansDine.totalsell) * 100
      y="{:.2f}".format(x)
      print(f"Branch consists of total sell's: {y}%")

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
